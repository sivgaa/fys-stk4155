```python
MNIST_logreg=np.loadtxt(save_path+'MNIST_logreg_acc_val.txt')
MNIST_FFNN=np.loadtxt(save_path+'MNIST_FFNN_5_acc_val.txt')
MNIST_CNN=np.loadtxt(save_path+'MNIST_model_30_modifisert_acc_val.txt')
FMNIST_logreg=np.loadtxt(save_path+'FashionMNIST_logreg_acc_val.txt')
FMNIST_FFNN=np.loadtxt(save_path+'FashionMNIST_FFNN_5_acc_val.txt')
FMNIST_CNN=np.loadtxt(save_path+'FashionMNIST_model_30_modifisert_acc_val.txt')

plt.plot(FMNIST_logreg,'--r',  label='Fashion MNIST LogReg')
plt.plot(FMNIST_FFNN,'--b', label='Fashion MNIST FFNN')
plt.plot(FMNIST_CNN,'--y', label='Fashion MNIST CNN')
plt.plot(MNIST_logreg,'r', label='MNIST LogReg')
plt.plot(MNIST_FFNN,'b', label='MNIST FFNN')
plt.plot(MNIST_CNN,'y', label='MNIST CNN')
plt.xlabel('Epoke')
plt.ylabel('Accuracy')
plt.legend(loc='upper left')
plt.savefig(save_path+'sammenligning_accuracy_val.png')
```
